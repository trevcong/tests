```python
# MATRIX MULTIPLICATION IN CODE
import numpy as np
# A                          At(a transpose)
#[1  -2  .1]       ->      [1    2]
#[2  -2  .2]               [-1  -2]
#                          [.1  .2]
A = np.array([[1, -1, 0.1], [2, -2, 0.2]]) 
AT = A.T # Transpose method 


# W
# [3  5  7  9]
# [4  6  8  0] 
W = np.array([[3, 5, 7, 9], [4, 6, 8, 0]]) 

# Z = A^tW                           or Z = AT @ W ( another way to say matmal )
#[11    17    23    9]
#[-11  -17  -23  -9]
#[1.1  1.7  2.3  0.9]

Z = np.matmul(AT, W)
print(Z)
```
